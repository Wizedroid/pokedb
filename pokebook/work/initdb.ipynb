## CONNECT TO THE DATABSE

```python
import psycopg2

connection = psycopg2.connect(dbname="pokemon", user="oak", password="professor", host="pokedb")
cursor = connection.cursor()
```

## Pokemon with the strongest move

```python
cursor.execute("SELECT * FROM pokemon JOIN pokemon_types ON pokemon.id = pokemon_types.pokemon_id JOIN types ON types.id = pokemon_types.type_id WHERE types.name = 'fire'")

column_names = [desc[0] for desc in cursor.description]
print(column_names)

for row in  cursor:
    print(row)
```

```python
cursor.execute("SELECT pokemon.name "
               "FROM pokemon "
               "WHERE pokemon.id IN ("
                   "SELECT pokemon_moves.pokemon_id FROM pokemon_moves JOIN moves ON moves.id = pokemon_moves.move_id "
                   "WHERE moves.power = (SELECT max(power) FROM moves) )")

column_names = [desc[0] for desc in cursor.description]
print(column_names)

for row in  cursor:
    print(row)
```

## Strongest Move
